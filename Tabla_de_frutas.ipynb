```python
from tabulate import tabulate

# Tabla de frutas (usamos una lista de diccionarios)
tabla_frutas = []

def agregar_fruta(nombre, valor):
    # Crear un diccionario con la fruta y su valor
    fruta = {'nombre': nombre, 'valor': valor}
    tabla_frutas.append(fruta)
    print(f"Fruta '{nombre}' agregada con éxito.\n")

def buscar_fruta(nombre):
    # Buscar la fruta por nombre en la tabla
    for fruta in tabla_frutas:
        if fruta['nombre'].lower() == nombre.lower():
            return fruta
    return None

def eliminar_fruta(nombre):
    # Buscar la fruta para eliminarla
    for fruta in tabla_frutas:
        if fruta['nombre'].lower() == nombre.lower():
            tabla_frutas.remove(fruta)
            print(f"Fruta '{nombre}' eliminada con éxito.\n")
            return
    print(f"Fruta '{nombre}' no encontrada.\n")

def mostrar_tabla():
    # Mostrar todas las frutas en la tabla de manera más bonita
    if len(tabla_frutas) == 0:
        print("No hay frutas en la tabla.\n")
    else:
        # Preparamos los datos con el número de caracteres en el nombre
        datos = [(len(fruta['nombre']), fruta['nombre'], fruta['valor']) for fruta in tabla_frutas]
        # Usamos tabulate para imprimir la tabla de forma ordenada
        print("\nTabla de frutas:")
        print(tabulate(datos, headers=["Índice", "Nombre", "Valor"], tablefmt="pretty"))
        print()  # Añadir una línea en blanco para mayor claridad

# Menú principal
def menu():
    while True:
        print("=== Menú ===")
        print("1. Agregar fruta")
        print("2. Buscar fruta")
        print("3. Eliminar fruta")
        print("4. Mostrar tabla de frutas")
        print("5. Salir")

        opcion = input("Seleccione una opción (1-5): ")

        if opcion == '1':
            nombre = input("Ingrese el nombre de la fruta: ")
            valor = input("Ingrese el valor de la fruta: ")
            try:
                valor = float(valor)
                agregar_fruta(nombre, valor)
            except ValueError:
                print("Error: El valor debe ser un número.\n")

        elif opcion == '2':
            nombre = input("Ingrese el nombre de la fruta a buscar: ")
            fruta = buscar_fruta(nombre)
            if fruta:
                print(f"Fruta encontrada: {fruta['nombre']} - Valor: {fruta['valor']}\n")
            else:
                print(f"Fruta '{nombre}' no encontrada.\n")

        elif opcion == '3':
            nombre = input("Ingrese el nombre de la fruta a eliminar: ")
            eliminar_fruta(nombre)

        elif opcion == '4':
            mostrar_tabla()

        elif opcion == '5':
            print("¡Hasta luego!\n")
            break

        else:
            print("Opción inválida, por favor seleccione una opción entre 1 y 5.\n")

# Ejecutar el menú
if __name__ == "__main__":
    menu()
```
