```python
import random

def radix_sort_negative(arr):
    """
    Ordena una lista de enteros negativos usando el algoritmo Radix Sort.

    Args:
        arr: Una lista de enteros negativos.

    Returns:
        La lista ordenada.
    """
    if not arr:
        return []

    # Convertir los números negativos a positivos y normalizar a 3 dígitos
    abs_arr = [abs(x) for x in arr]

    # Número máximo de dígitos (3, ya que los valores son hasta -999)
    max_digits = 3

    # Radix Sort
    for i in range(max_digits):
        buckets = [[] for _ in range(10)]  # Crear cubetas del 0 al 9
        for num in abs_arr:
            digit = (num // (10 ** i)) % 10  # Extraer el dígito correspondiente
            buckets[digit].append(num)

        # Reconstruir la lista
        abs_arr = [num for bucket in buckets for num in bucket]

    # Convertir los números positivos de vuelta a negativos
    return [-num for num in abs_arr]


def radix_sort(arr):
    """
    Ordena una lista de enteros positivos usando el algoritmo Radix Sort.

    Args:
        arr: Una lista de enteros positivos.

    Returns:
        La lista ordenada.
    """
    if not arr:
        return []

    # Encontrar el número con más dígitos
    max_num = max(arr)
    num_digits = len(str(max_num))

    # Radix Sort
    for i in range(num_digits):
        buckets = [[] for _ in range(10)]  # Crear cubetas del 0 al 9
        for num in arr:
            digit = (num // (10 ** i)) % 10  # Extraer el dígito correspondiente
            buckets[digit].append(num)

        # Reconstruir la lista
        arr = [num for bucket in buckets for num in bucket]

    return arr


def menu():
    print("\nSeleccione una opción:")
    print("1. Ordenar números negativos usando Radix Sort")
    print("2. Ordenar números positivos usando Radix Sort")
    print("3. Salir")
    return input("Ingrese el número de la opción: ")


if __name__ == "__main__":
    while True:
        option = menu()
        if option == "1":
            num_elements = int(input("Ingrese el número de elementos: "))
            negative_numbers = [random.randint(-999, -1) for _ in range(num_elements)]
            print(f"Lista no ordenada: {negative_numbers}")
            sorted_numbers = radix_sort_negative(negative_numbers)
            print(f"Lista ordenada: {sorted_numbers}")

        elif option == "2":
            num_elements = int(input("Ingrese el número de elementos: "))
            positive_numbers = [random.randint(1, 999) for _ in range(num_elements)]
            print(f"Lista no ordenada: {positive_numbers}")
            sorted_numbers = radix_sort(positive_numbers)
            print(f"Lista ordenada: {sorted_numbers}")

        elif option == "3":
            print("Saliendo...")
            break

        else:
            print("Opción no válida, por favor intente de nuevo.")
```
